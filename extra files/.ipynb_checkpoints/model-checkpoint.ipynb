```python
import keras 
import pandas as pd
import matplotlib.pyplot as plt 
import keras 
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image 
import os
import numpy as np
import math
import dlib
import cv2
import easydict
```

```python
args={"shape_predictor":"shape_predictor_68_face_landmarks.dat"}
```

```python
detector=dlib.get_frontal_face_detector
predictor=dlib.shape_predictor(args["shape_predictor"])
```

```python
def rect_to_bb(rect):
    x=rect.left()
    y=rect.top()
    w=rect.right()-x
    h=rect.bottom()-y
    return (x,y,w,h)
```

```python
def shape_to_np(shape,dtype='int'):
    coords=np.zeros((68,2),dtype=dtype)
    for i in range(0,68):
        coords[i]=(shape.part(i).x,shape.part(i).y)
    return coords
```

```python
def preprocess_dis(dis_array):
    x=dis_array-np.mean(dis_array,axis=0)
    x/=np.std(dis_array,axis=0)
    return x
```

```python
def convert_points_to_distance(points):
    mid_x=(points[:,14,0]+points[:,18,0])/2
    mid_y=(points[62,1]+points[66,1])/2
    mid_point=np.array([mid_x,mid_y])
    points=(points-mid_point)**2
    distances_array=np.sqrt((points[:,:,0]+points[:,:,1]))
    
    return preprocess_dis(distances_array)
    
```

```python
def face_landmarks(images):
    all_images=[]
    for image in images:
        gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
        rects=detector(gray,1)
        
        for rect in rects:
            shape=predictor(gray,rect)
            shape=shape_to_np(shape)
            shape=shape[48:68]
        all_images.append(shape)
    
    return all_images    
```

```python
x=os.listdir("words")
```

```python
labels=[]
lip_data=[]
datagen=ImageDataGenerator()
train_dir="words"
for label in os.listdir("words"):
    word_dir=os.path.join(train_dir,label)
    for repeat in os.listdir(word_dir):
        image_dir=os.path.join(word_dir,repeat)
        
        print(image_dir)
        image_dir="words/01/01"
        for batch in datagen.flow_from_directory(directory=image_dir,batch_size=10,shuffle=False):
            points=face_landmarks(batch[0])
            lip_data.append(convert_points_to_distance(points))
            labels.append(int(label)-1)
            
```
